### List Comprehension
List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]


Advantages of List Comprehension

More time-efficient and space-efficient than loops.

Require fewer lines of code.

Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list (normal code)

L =[]
for i in range(1,11):
    L.append(i)
print(L)
```

```python
# Add 1 to 10 numbers to a list (list comprehension)

LC = [i for i in range(1,11)]
LC
```

```python
# scalar multiplication on a vector in a list (normal code)
L =[]
v = [2,3,4]
s = -5

for i in v:
    result =i*s
    L.append(result)
    
print(L)
```

```python
# scalar multiplication on a vector in a list (list comprehension)

v = [2,3,4]
s = -5

LC = [i*s for i in v]
LC
```

```python
# Add squares normal code
L=[] 
num = [1,2,3,4]

for i in num:
    result = i**2
    L.append(result)
print(L)
    
    
```

```python
# Add squares List comprehension
 
num = [1,2,3,4,5]

LC = [i**2 for i in num]
LC
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50

LC = [i for i in range(1,51) if i%5 == 0]
LC
```

```python
# find languages which start with letter p

languages = ['java','python','php','c','javascript']

LC = [language for language in languages if language.startswith('p')]
LC
```
