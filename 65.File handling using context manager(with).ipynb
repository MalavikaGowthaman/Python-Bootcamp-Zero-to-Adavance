### File Handling using Context Manager (With)

#### Using Context Manager (With)

- It's a good idea to close a file after usage as it will free up the resources
- If we dont close it, garbage collector would close it
- with keyword closes the file as soon as the usage is over

#####  no need to use f.close() in this method this is the method need to use in project

#### write()

```python
with open('files/sample.txt','w') as f:
    f.write("Hello there!")
    
```

```python
f.write("Hello every one") # it gets automatically closed no need to use close function
```

```python
with open('files/sample.txt','w') as f:
    f.write("Hello there!")
    f.write("\nHello every one")
```

#### read()

```python
with open('files/sample.txt','r') as f:
    string =f.read()
    print(string)
```

#### readline()

```python
with open('files/sample.txt','r') as f:
    string =f.readline()
    print(string)
```

#### append

```python
with open('files/sample.txt','a') as f:
    f.write("\nWelcome to python world")
    
```

#### to load a big data which hold high memory can able to upload the data by chuncks

```python
big_txt = ['Welcome to python world ' for i in range(10000)]

with open('files/big.txt','w') as f:
    f.writelines(big_txt)
```

```python
with open('files/big.txt','r') as f:
    chunck_size = 10
    
    while len(f.read(chunck_size)) > 0:
        print(f.read(chunck_size),end ='')
        f.read(chunck_size)
```

```python

# read()
with open("files/big.txt", "r") as f:
    s = f.read()
    print(s)
```

```python

# seek and tell function

with open("files/sample.txt", "r") as f:
    print(f.read(10))
    print(f.read(10))
    print(f.tell())
```

```python
with open("files/sample.txt", "r") as f:
    
    print(f.read(10))
    f.seek(0)
    print(f.tell())
```

```python
with open("files/sample.txt", "w") as f:
    f.write("Welcome")
    f.seek(0)
    f.write("Da")
    print(f.tell())
```
