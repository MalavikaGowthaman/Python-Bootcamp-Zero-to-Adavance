### Serialization and Deserialization

* **Serialization** - process of converting python data types to JSON format
* **Deserialization** - process of converting JSON to python data types

What is JSON?
The full form of JSON is JavaScript Object Notation.



```python
# serialization 

# list
import json
L = [1,2,3,4,5,6]

with open('files/test.json','w') as f:
    json.dump(L,f)
```

```python
# dict

import json
d = {
    
    'name': 'Malavika',
    'age': 26,
    'gender':'female'
    
}

with open('files/test.json','w') as f:
    json.dump(d,f)
```

```python
# dict

import json
d = {
    
    'name': 'Malavika',
    'age': 26,
    'gender':'female'
    
}

with open('files/test.json','w') as f:
    json.dump(d,f,indent=3) # get space in json file
```

```python
# tuple 
import json
t = (1,2,3,4,5,6)

with open('files/test.json','w') as f:
    json.dump(t,f)
```

```python
# tuple 
import json
s = {1,2,3,4,5,6}

with open('files/test.json','w') as f:
    json.dump(s,f) # set can't get serialize
```

```python
# deserialization

import json
with open('files/test.json','r') as f:
    data = json.load(f)
    print(data)
    print(type(data))
```
