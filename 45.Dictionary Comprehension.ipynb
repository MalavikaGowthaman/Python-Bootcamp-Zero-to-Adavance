### Dictionary Comprehension



```python
# print 1st 10 numbers and their squares

for i in range(1,11):
    print(i,':',i**2)
    
```

```python
# print 1st 10 numbers and their squares using dict comprehension

DC = {i: i**2 for i in range(1,11)}
DC
```

```python
# using if condition
products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

for key,value in products.items():
    if value >0 :
        print(key,':',value)
```

```python
# using dict comprehesion

products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

DC = {key : value for key,value in products.items() if value > 0}
DC
```

```python
products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

DC = {key : value for key,value in products.items() if value > 0}

for key,value in DC.items():
    print(f"{key} = {value}")
```

```python
# using zip
days = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

for i,j in zip(days,temp_C):
    print(i,j)
```

```python
DC = {i :  j for i,j in zip(days,temp_C)}
DC
```

```python
# using zip
days = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

DC = {i:j for i,j in zip(days,temp_C)}
for i,j in DC.items():
    print(f"{i} = {j}")
```

```python
days = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

DC = [(i,j) for i,j in zip(days,temp_C)]
DC
for i,j in DC:
    print(f"{i} = {j}")
```
