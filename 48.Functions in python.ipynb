```python
def greeting(name):
    output = f"Hello {name}, Welcome to python"
    return(output)
    
greeting("Malavika")  
```

```python
names = ['Amar','Alex','Hari','Gowri','Ruba']

for i in names:
    output = f"Hello {i}, Welcome to python"
    print(output)
```

```python
for name in names:
    output = greeting(name) # can able to use the function here
    print(output)
```

```python
def odd_even(num):
    if num % 2 ==0:
        return("Even number")
        
    else:
        return("Odd number")

odd_even(num = int(input("Enter number : ")))
```

```python
for i in range(1,11):
    result = odd_even(i)
    print(i ,"=", result)
```
